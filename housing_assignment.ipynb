## Encoding

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
```

```python
label_encoder = LabelEncoder()
data['ID_enc'] = label_encoder.fit_transform(data['ID'])
data
```

```python
data.info()
```

```python
category_val = ['Waterfront View','Condition of the House']
data_enc= pd.get_dummies(data, columns=category_val,dtype=int,drop_first=True)
data_enc
```

##Scaling
standard normal

```python
scale_column=['Sale Price','No of Bathrooms','Flat Area (in Sqft)','Lot Area (in Sqft)','Area of the House from Basement (in Sqft)','Living Area after Renovation (in Sqft)','Lot Area after Renovation (in Sqft)']
```

```python
data_scal = data_enc.copy()

std_scaler = StandardScaler()
data_scal[scale_column] = std_scaler.fit_transform(data_scal[scale_column])
data_scal.head()
```

minmax scaler

```python
minmax_scal = data_enc.copy()

minmax_scaler = MinMaxScaler()
minmax_scal[scale_column] = minmax_scaler.fit_transform(minmax_scal[scale_column])
minmax_scal.head()
```

```python
minmax_scal.describe()
```
